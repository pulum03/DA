```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

raw = sns.load_dataset('titanic')
raw.corr()
```

```python
df = raw.corr()
df
```

```python
df = raw.corr()

sns.clustermap(df, 
               annot = True,      #    
               cmap = 'RdYlBu_r',  # Red, Yellow, Blue  
               vmin = -1, vmax = 1, # -1 ~ 1  
              )
```

```python
df = raw.corr()
#   
fig, ax = plt.subplots( figsize=(7,7) )

#   (   True,   False)
mask = np.zeros_like(df, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

#  
sns.heatmap(df, 
            cmap = 'RdYlBu_r', 
            annot = True,   #   
            mask=mask,      #     
            linewidths=.5,  #   
            cbar_kws={"shrink": .5},#    
            vmin = -1,vmax = 1   #   -1 ~ 1
           )  
plt.show()
```
