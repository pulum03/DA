```python
# data from Kaggl's home credit default risk
import pandas as pd
data = pd.read_csv('home-credit-default-risk/application_train.csv')
```

```python
# 컬럼별 type 확인 및 결측치 확인
data.info()
data.isnull().sum()
```

```python
print(data[:5])
```

```python
# 문자 전환
data = data.replace(' ', '')
```

```python
# 만약 결측치가 문자열 스페이스(' ')로 되어있다면, np.nan으로 바꿔 pandas 라이브라리가 인식할 수 있도록 변환
import numpy as np
data = data.replace('', np.nan)
```

```python
# 결측치 row 제거
data.dropna(how='all')  # 'all' : 한행이 모두 nan이면 제거, 'any' : 한행이라도
```

```python
# 결측치 처리
data.fillna(0, inplace = True)
```

```python
data.isnull().sum()
```

```python
# 남자/여자, 소형/중형/대형 등 문자로 표현되는 범주형 데이터는 One & Hot 인코딩으로 처리된다.
# One & Hot 인코딩 을 하게 되면 특정 변수에서 나올수 있는 문자열 수대로 새로운 항목이 생성 된다.
# 소형/중형/대형 데이터를 같는 변수의 인코딩 처리시 소형여부,중형여부, 대형여부 총 3가지 항목값
# 값이 중형일 경우 소형여부, 중형여부, 대형여부의 매핑값은 0, 1, 

# 남자/여자처럼 0,1 두가지 코드로 해결할 수 있는 경우
# 한 항목을 사용해 0,1로 표현할 수 있으므로 유사변수는 category_list로 따로 분류

# 처리방식에 따른 컬럼 정리
category_list = ['NAME_CONTRACT_TYPE', 'FLAG_OWN_CAR', 'FLAG_OWN_REALTY']
one_hot_list = ['NAME_TYPE_SUITE', 'NAME_INCOME_TYPE', 'NAME_EDUCATION_TYPE', 'NAME_FAMILY_STATUS',
               'NAME_HOUSING_TYPE', 'OCCUPATION_TYPE', 'WEEKDAY_APPR_PROCESS_START', 'ORGANIZATION_TYPE',
               'FONDKAPREMONT_MODE', 'HOUSETYPE_MODE', 'WALLSMATERIAL_MODE', 'CODE_GENDER', 'EMERGENCYSTATE_MODE']
```

```python
# categorize
categories_encoded = pd.DataFrame()
cate_cols = []

for x in category_list:
    #print(x)
    X = data[x]
    #print(X)
    x_encoded, x_categories = X.factorize()
    #print(x_encoded[:5], x_categories[:5])
    
    # dataframe
    temp_df = pd.DataFrame(x_encoded)
    #print(temp_df[:5])
    categories_encoded = pd.concat([categories_encoded, temp_df], axis = 1) # 행으로 붙이기
        
    # 컬럼명 추가
    cate_cols.append(x + '_1')
    
# 컬럼명 수정
categories_encoded.columns = [cate_cols]

print(categories_encoded.head())
```

```python
# 항목별로 max값을 확인하여 0,1이상의 값을 갖는지 확인. 모두 1이면 정상, 2이상나오는 값은 one_hot_list로 넘김

# 항목별 맥스값 체크(확인용)
for x in category_list:
    col = x + '_1'
    print(x, 'max : ', max(categories_encoded[col].values)[0])
```

```python
# reshape 예시
x = np.arange(12)
print(x)
x = x.reshape(3,4)
print(x)

x.reshape(-1,1) # shape(12,1)
x.reshape(-1,2) # shape(6,2)
x.reshape(-1,3) # shape(4,3)
x.reshape(-1,4) # shape(3,4)

x.reshape(1,-1) # shape(1,12)
x.reshape(2,-1) # shape(2,6)
x.reshape(3,-1) # shape(3,4)
x.reshape(4,-1) # shape(4,3)
```

```python
# one_hot_list 처리
from sklearn.preprocessing import OneHotEncoder
# One_Hot_Encoder
encoder = OneHotEncoder()
onehot_encoded = pd.DataFrame()
onehot_cols = []

for x in one_hot_list:
    #print(x)
    X = data[x]
    x_encoded, x_categories = X.factorize()
    
    #print(x_encoded)
    #print(x_encoded.reshape(-1,1))
    
    x_1hot = encoder.fit_transform(x_encoded.reshape(-1,1))
    x_1hot = x_1hot.toarray()
    #print(x_1hot)
    
    # dataframe
    temp_df = pd.DataFrame(x_1hot)
    onehot_encoded = pd.concat([onehot_encoded, temp_df], axis =1)
    #print(onehot_encoded.head())
    
    # 컬럼명 추가
    for i in range(1, temp_df.shape[1] + 1):
        onehot_cols.append(x + '_' + str(i))
        
# 컬럼명 수정
onehot_encoded.columns = [onehot_cols]
#print(onehot_encoded.head())
```

```python
total_cols = set(data.columns.values)
#print(total_cols)
final_cols = list(total_cols - set(category_list) - set(one_hot_list))
#print(final_cols)
```

```python
final_df = pd.concat([categories_encoded, onehot_encoded, data[final_cols]], axis = 1)
```

```python
print('최초컬럼 -> 최종컬럼')
print(len(total_cols), len(final_cols))

print('병합 전 3개 테이블')
print(categories_encoded.shape, onehot_encoded.shape, data[final_cols].shape)

print('병합 후 1개 테이블')
final_df.shape
```

```python
# 컬러명 재지정
final_df.columns = cate_cols + onehot_cols + final_cols
final_df.head()
```

```python
final_df.to_csv('home-credit-default-risk/home_credit_risk.csv')
```

```python
import pickle

with open('home_credit_risk.p', 'wb') as file: # hello.txt 파일을 바이너리 쓰기 모드
    pickle.dump(final_df, file)
```
