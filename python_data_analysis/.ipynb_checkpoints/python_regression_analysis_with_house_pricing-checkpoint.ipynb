```python
import numpy as np
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
print(boston)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
# x: average number of rooms per dwelling(방 개수), y: 주택 가격
plt.scatter(boston.data[:, 5], boston.target, color='r', s=10)
print(boston.DESCR)
```

```python
# 1차원 회귀
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

x = boston.data[:,5]
print(x)
# LinearRegression의 함수에 인자로 전달하기 위해서는
# 각 원소의 값이 담긴 2차원 배열을 (n, 1)의 형태로 만들어야 합니다.
# atleast_2d 함수는 x 데이터를 2차원 배열을 (1, n) 형태로 만들어줍니다.
print('')
x_1 = np.atleast_2d(x)
print(x_1)

print('')
# transpose 함수는 (1,n) -> (n,1)
x_2 = np.transpose(x_1)
print(x_2)

y = boston.target

# fit 함수로 간단하게 회귀모형
lr.fit(x_2,y)
```

```python
%matplotlib inline
y_predicted = lr.predict(x_2)
plt.plot(x_2, y_predicted)
```

```python
%matplotlib inline
plt.scatter(boston.data[:, 5], boston.target, color='r', s=10)
plt.plot(x_2, y_predicted)
```
