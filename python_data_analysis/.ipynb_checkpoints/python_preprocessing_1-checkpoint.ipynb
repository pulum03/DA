```python
import pandas as pd
```

```python
# 전체적인 개요 살피기

# data import
data = pd.read_csv('titanic/train.csv')
print(data.describe())

# Age column의 count 가 다른 column 보다 갯수가 적음을 확인할 수 있습니다. 
# 즉, NaN값이 존재한다는 의미가 될 수 있겠습니다.

# Survived의 column과 같은 경우, 탑승객이 살았다면 1 그렇지 않다면 0으로 표기되어 있습니다. 
# mean은 전체 데이터의 합을 갯수로 나눈 평균값이므로, 
# Survived의 mean은 곧 전체 탑승객의 생존율을 말해줍니다.
```

```python
print(data.info())

# info()의 메소드를 활용해서도 NaN값이 어느 column에서 빠졌는지 쉽게 확인할 수 있습니다. 
# Age, Cabin, Embarked에서 빠졌음을 볼 수 있습니다. 
# Embarked column은 단 2곳만이 빠졌기 때문에 눈으로 일일이 확인했더라면 놓쳤을 수도 있었겠습니다.
```

```python
from sklearn.preprocessing import Imputer
```

```python
df = data.copy()
# text로 표기된 column은 drop
df.drop(['Name', 'Sex', 'Ticket', 'Cabin', 'Embarked'], axis=1, inplace=True)
df.head(10)
```

```python
imputer = Imputer(strategy='median')
# strategy를 옵션 값으로 주게 되어있는데, 
# strategy의 옵션 값으로는 ‘mean’, ‘median’, ‘most_frequent’

imputer.fit(df)
print(imputer.statistics_)
# 데이터를 fit한 후에 statistics_에 저장된 결과 값을 출력해 볼 수 있습니다.
```

```python
output = imputer.transform(df)

# DataFrame으로 전환
output = pd.DataFrame(output, columns=df.columns, index=list(df.index.values))
print(output.info())

# 모든 값들이 채워짐
```

```python
# 기존의 성별 column 인 data[‘Sex’]를 출력해 보면 
# ‘male’과 ‘female’로 구성되어 있음을 알 수 있습니다. 
# 성별은 binary 속성 (남 / 여) 으로 구성되어 있기 때문에 
# 이를 0 / 1로 변환해 주고 싶다면 factorize() 를 호출하여 손쉽게 변환
```

```python
sex_encoded, sex_category = data['Sex'].factorize()
print(sex_encoded)
print('')
print(sex_category)
```

```python
# sklearn.preprocessing 패키지를 활용하여 one-hot encoding
# ‘Embarked’ column을 살펴 보면, ‘C’, ‘S’, ‘Q’ 3가지로 이뤄져 있음을 확인

# NaN값에 'S' 입력 처리
data.loc[data['Embarked'].isnull(), 'Embarked'] = 'S'

from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder()
embarked_encoded, embarked_categories = data['Embarked'].factorize()
print(embarked_encoded)
print(embarked_categories)

embarked_hot = encoder.fit_transform(embarked_encoded.reshape(-1,1))
print(embarked_hot.toarray())
```

```python
# sklearn.preprocessing 패키지가 제공하는 MinMaxScaler / StandardScaler를 
# 활용하여 데이터 정규화

# min-max scaling 
def norm(x):
    _max = x.max()
    _min = x.min()
    _denom = _max - _min
    return (x - _min) / _denom

# sklearn 이용
from sklearn.preprocessing import MinMaxScaler
#print(output.head())
min_max_scaler = MinMaxScaler()
fitted =  min_max_scaler.fit(output)
print(fitted.data_max_)

print('')
output = min_max_scaler.transform(output)
print(output)

print('')
output = pd.DataFrame(output, columns=df.columns, index=list(df.index.values))
print(output.head())
```

```python
# standardization
from sklearn.preprocessing import StandardScaler
std_scaler = StandardScaler()
output.head()

fitted = std_scaler.fit(output)
print(fitted.mean_)

print('')
output = std_scaler.transform(output)
print(output)

print('')
output = pd.DataFrame(output, columns=df.columns, index=list(df.index.values))
print(output.head())
```
