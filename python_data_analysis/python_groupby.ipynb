```python
df = pd.read_csv('census.csv')
df = df[df['SUMLEV']==50]
```

```python
# 인덱스를 STNAME으로 한 후, fun함수를 groupby에 인자로 넣어 데이터 추출
# 이 때, fun에 들어가는 item 인수는 인덱스 임.
df = df.set_index('STNAME')

print(df)

def fun(item):
    if item[0] < 'M':
        return 0
    if item[0] < 'Q':
        return 1
    return 2

for group, frame in df.groupby(fun):
    print('There are ' + str(len(frame)) + ' records in group ' + str(group) + ' for processing')
```

```python
df = pd.read_csv('census.csv')
df = df[df['SUMLEV']==50]
```

```python
# STNAME을 통해 집계 후, CENSUS2010POP 컬럼을 기준으로 avg 집계
df_avg = df.groupby('STNAME').agg({'CENSUS2010POP' : np.average})
print(df_avg)
```

```python
# STNAME으로 인덱싱. level = 0 은 멀티인덱스일 시 가장 첫번째 인덱스 기준으로 집계하라는 의미
# agg함수에 전체 평균과 합을 동시에 쓸 수 있다.
census_avg_sum = (df.set_index('STNAME').groupby(level=0)['CENSUS2010POP']
                 .agg({'avg':np.average, 'sum':np.sum}))

print(census_avg_sum)
```

```python
pop_avg_sum = df.set_index('STNAME').groupby(level=0)['POPESTIMATE2010', 'POPESTIMATE2011']\
                .agg({'avg':np.average, 'sum':np.sum})
print(pop_avg_sum)
```
